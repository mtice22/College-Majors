# Visualizations 
Unlike the original Five Thirty-Eight article, I wanted to create some basic scatter plots to see if there is a correlation between certain features. 
## Percentage of Women vs Median Income
The most obvious relationship to look at is if gender distribution in each major affects median income. While there are many reasons for this, there is a clear negative linear relationship that can be seen. The outlier is Petrolueum Engineering.  

```python
plt.scatter(data['sharewomen'], data['median'])
plt.xlabel('% of Women in Major')
plt.ylabel('Median Income');
```

## Percentage of Women vs Unemployment Rate
After seeing the first relationship, I wanted to test if the percentage of women correlates with any other features. I compared % of women and unemployment rate but found no correlation. 

```python
plt.scatter(data['sharewomen'], data['unemployment_rate'])
plt.xlabel('% of Women in Major')
plt.ylabel('Unemployment Rate')
```

## Percentage of Women vs Percentage of Jobs not Requiring a Degree
Once again, no correlation is found. 

```python
plt.scatter(data['sharewomen'], data['rate_noncollege_jobs'])
plt.xlabel('% of Women in Major')
plt.ylabel('Percentage of Jobs not Requiring Degree')
```

# Top 10 Median Salaries by Women-Dominated Majors
By placing the filter sharewomen > 0.5, we can find the women-dominated majors that have the highest median salary. 

```python
d1 = data[['major','median','sharewomen']]
filtered_data = d1[d1['sharewomen']>0.5]
filtered_data.head(10)
```

# Top 10 Highest Unemployment Rates by Major

```python
d2 = data[['major', 'unemployment_rate']]
result = d2.sort_values(['unemployment_rate', 'major'], ascending=[0,1])
result.head(10)
```
